```python
import chromedriver_binary
from bs4 import BeautifulSoup
from selenium import webdriver
import numpy as np
import pandas as pd 
import time
from tqdm import tqdm
```

```python
browser=webdriver.Chrome()
browser.get("https://www.youtube.com/@GeeksforGeeksVideos/videos")
time.sleep(20)
soup=BeautifulSoup(browser.page_source,'html.parser')
```

```python
data=[]
for sp in tqdm(soup.find_all(id='dismissible')):
    video_link="https://www.youtube.com"+sp.find(id='thumbnail',class_='yt-simple-endpoint inline-block style-scope ytd-thumbnail').get('href')
    try:
        image=sp.find('img',class_='yt-core-image yt-core-image--fill-parent-height yt-core-image--fill-parent-width yt-core-image--content-mode-scale-aspect-fill yt-core-image--loaded').get('src').split('?')[0]
    except:
        image=np.nan
    title=sp.find(id='video-title-link',class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").get('title')
    views=sp.find(class_='inline-metadata-item style-scope ytd-video-meta-block').text
    date_time=sp.find_all(class_='inline-metadata-item style-scope ytd-video-meta-block')[1].text
    data.append([title,date_time,views,image,video_link])
```

```python
df=pd.DataFrame(data,columns=['title','date_time','views','image','video_link'])
df.to_csv('youtubr_scapping_1')
```

```python
browser = webdriver.Chrome()
browser.get('https://www.youtube.com')
time.sleep(2)
data = []
for link in tqdm(df['video_link']):
    browser.get(link)
    time.sleep(5)
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    try:
        title = soup.find_all('yt-formatted-string', class_ = 'style-scope ytd-video-primary-info-renderer')[0].text
    except:
        title = np.nan
    try:
        view = soup.find('span' , class_ = 'view-count style-scope ytd-video-view-count-renderer').text
    except:
        view = np.nan
    try:
        date_time = soup.find_all('yt-formatted-string', class_ = 'style-scope ytd-video-primary-info-renderer')[1].text
    except:
        date_time = np.nan
    try:
        like_divs = soup.find_all('div', class_='yt-spec-button-shape-next__button-text-content')
        like = np.nan
        for div in like_divs:
            text = div.text.strip()
            if text.replace('.', '').replace('K', '').replace('M', '').isdigit():
                like = text
                break
    except:
        like = np.nan
    try:
        desc_box = soup.find('div', id='description')
        desc_tag = desc_box.find('yt-formatted-string') if desc_box else None
        description = desc_tag.text.strip() if desc_tag and desc_tag.text.strip() else np.nan
    except:
        description = np.nan
    data.append([title , date_time, like, view, link, description])
```

```python
df=pd.DataFrame(data,columns=['title','date_time','like','view','likn','description'])
df.to_csv('youtube_scrapping_2')
```
